These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

```python

best_math_schools = schools[schools["average_math"] >= 640][["school_name", "average_math"]]
best_math_schools = best_math_schools.sort_values(by="average_math",ascending = False)
```

```python
schools["total_SAT"] = schools["average_math"]+schools["average_reading"]+schools["average_writing"]
schools_sorted = schools.sort_values(by="total_SAT", ascending = False)
top_10_schools = schools_sorted.head(10)[["school_name","total_SAT"]]
```

```python
grouped = schools.groupby("borough")["total_SAT"].agg(['std', 'mean', 'count'])

# Найти район с наибольшим стандартным отклонением
max_std_borough = grouped['std'].idxmax()
max_std_values = grouped.loc[max_std_borough]

# Создание DataFrame с результатами
largest_std_dev = pd.DataFrame({
    "borough": [max_std_borough],
    "num_schools": [max_std_values['count']],
    "average_SAT": [max_std_values['mean']],
    "std_SAT": [max_std_values['std']]
})

# Округление всех числовых значений до двух десятичных знаков
largest_std_dev = largest_std_dev.round(2)
```
